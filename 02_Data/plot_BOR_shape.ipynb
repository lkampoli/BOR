```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
```

```python
matplotlib.rc('xtick', labelsize=25) 
matplotlib.rc('ytick', labelsize=25)

font = {'family' : 'sans-serif',
        'weight' : 'bold',
        'size'   : 25}

matplotlib.rc('font', **font)
```

```python
BOR = pd.read_csv('BOR_shape.csv')
BOR_full = pd.read_csv('BOR_shape_full.csv')
```

```python
plt.figure(figsize=(30,15), frameon=False)
plt.axhline(y = 0, xmin=0.05, xmax=0.19, color = 'lightgrey', linewidth=5)
plt.axhline(y = 0, xmin=0.67, xmax=0.95, color = 'lightgrey', linewidth=5)
plt.scatter(BOR_full['C:0'], BOR_full['C:1'], label='BOR', linewidth=2, c='lightgrey', linestyle = '-')


plt.ylabel("$y/R$ [-]")
plt.xlabel("$U/u_{b}$ [-]")
plt.tight_layout()
#plt.legend(['top', 'bottom', 'LES', 'kOmegaSST-BSL', 'kOmegaSST-GEP'], loc='upper left')
plt.savefig('BOR_domain.pdf')
plt.grid()
#plt.legend(['top', 'bottom', 'LES', 'Fabian BSL', 'Fabian best', 'kOmegaSST', 'kEpsilon', 'kEpsilonPhitF', 'RNGkEpsilon'], loc='lower right')
#plt.legend(['top', 'bottom', 'LES', 'Fabian BSL', 'Fabian best', 'kOmegaSST', 'kEpsilon', 'kEpsilonPhitF'], loc='upper left')
plt.show()
plt.close()
```
