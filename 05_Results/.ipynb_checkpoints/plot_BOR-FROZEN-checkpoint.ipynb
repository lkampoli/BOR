```python
sol_BOR_U = []
sol_BOR_R = []

for r in range(0,len(BOR)):
    #print("Region =", PART[r])
    #print("=================")
    
    sol_U = []
    sol_R = []

    for s in range(0,len(solution_dir)):
        #print("   Solution =", solution_name[s])
        
        sol_profile_U = []    
        sol_profile_R = []
                
        for p in range(0, len(BOR[r])):
            f = BOR[r][p]
            #print("      Profile =", f)
            try:
                
                if( (solution_name[s]=='SST_MediumBody-GEP_aijxOFF_RijhatON') or 
                    (solution_name[s]=='SST_MediumBody-GEP_aijxON_RijhatON') 
                  ):
                    
                    if(f==-0.1):
                        sol_profile_U.append(np.loadtxt(solution_dir[s]+'XoRm0.1_U.xy')) 
                        sol_profile_R.append(np.loadtxt(solution_dir[s]+'XoRm0.1_R.xy')) 
                    elif(f==0.0):
                        sol_profile_U.append(np.loadtxt(solution_dir[s]+'XoR0.0_U.xy')) 
                        sol_profile_R.append(np.loadtxt(solution_dir[s]+'XoR0.0_R.xy')) 
                    else:
                        sol_profile_U.append(np.loadtxt(solution_dir[s]+'XoRp'+str(f)+'_U.xy'))
                        sol_profile_R.append(np.loadtxt(solution_dir[s]+'XoRp'+str(f)+'_R.xy'))
                
                else:
                
                    if(f==-0.1):
                        sol_profile_U.append(np.loadtxt(solution_dir[s]+'XoRm0.1_U.xy')) 
                        sol_profile_R.append(np.loadtxt(solution_dir[s]+'XoRm0.1_Rall.xy')) 
                    elif(f==0.0):
                        sol_profile_U.append(np.loadtxt(solution_dir[s]+'XoR0.0_U.xy')) 
                        sol_profile_R.append(np.loadtxt(solution_dir[s]+'XoR0.0_Rall.xy')) 
                    else:
                        sol_profile_U.append(np.loadtxt(solution_dir[s]+'XoRp'+str(f)+'_U.xy'))
                        sol_profile_R.append(np.loadtxt(solution_dir[s]+'XoRp'+str(f)+'_Rall.xy'))
                                            
            except:
                pass
                #print("         skip:", p, "-->", f)
            
        sol_U.append(sol_profile_U)
        sol_R.append(sol_profile_R)
        
    sol_BOR_U.append(sol_U)
    sol_BOR_R.append(sol_R)    
```
